```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv("/content/tmdb_5000_movies.csv")
credits=pd.read_csv("/content/tmdb_5000_credits.csv")
```

```python
movies=movies.merge(credits, on="title")
```

```python
movies.columns
```

```python
movies=movies[['movie_id', 'genres', 'keywords', 'overview', 'title', 'cast',  'crew']]
```

```python
movies.head(2)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.iloc[0].genres
```

```python
import ast
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head(2)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert3(obj):
  L=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter!=3:
      L.append(i['name'])
      counter+=1
    else:
        break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies['crew'][0]
```

```python
import ast
def fetch_director(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head(2)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x] if x is not None else [])
```

```python
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies['tag'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df=movies[['movie_id', 'title', 'tag']]
```

```python
new_df.head(2)
```

```python
new_df['tag']=new_df['tag'].apply(lambda x:" ".join (x))
```

```python
movies.head()
```

```python
new_df['tag']=new_df['tag'].apply(lambda x:x.lower())
```

```python
new_df['tag'][0]
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tag']=new_df['tag'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tag']).toarray()
```

```python
vectors
```

```python
len(cv.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity
```

```python
import warnings
warnings.filterwarnings('ignore', category=UserWarning)

def recommend(movie):
  movie_lower = movie.lower() # Convert input movie title to lowercase for case-insensitive matching

  # Check if the lowercase movie title exists in the lowercase version of the DataFrame's title column
  if movie_lower not in new_df['title'].str.lower().values:
    print("Sorry, no any such movie")
    return

  # Find the index of the movie using the lowercase comparison
  movie_index = new_df[new_df['title'].str.lower() == movie_lower].index[0]

  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]

  print(f"Recommendations for '{movie}':") # Print the original capitalized movie name for better display
  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend("Batman")
```

```python
recommend("SuperMan")
```
